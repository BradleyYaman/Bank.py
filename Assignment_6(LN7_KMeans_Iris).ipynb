```python
# Import the required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris  # Iris dataset
import os
os.environ["OMP_NUM_THREADS"] = "1"
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Updated K values
K_values_updated = [5, 6, 7]

# Plot results for each K
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, K in enumerate(K_values_updated):
    kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)

    # Plot clustering result
    axes[i].scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

    # Plot cluster centers
    centers = kmeans.cluster_centers_
    axes[i].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)

    axes[i].set_xlabel(iris.feature_names[0])
    axes[i].set_ylabel(iris.feature_names[1])
    axes[i].set_title(f'K-Means Clustering (K={K})')

plt.tight_layout()
plt.show()
```
